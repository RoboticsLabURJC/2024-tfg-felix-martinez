```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

# Rango de valores para x
x = torch.linspace(-5, 5, steps=1000)

# Funciones de activacin
gelu = F.gelu(x)
relu = F.relu(x)
swish = x * torch.sigmoid(x)  # Swish = x * sigmoid(x)

# Convertir a numpy para graficar
x_np = x.numpy()
gelu_np = gelu.numpy()
relu_np = relu.numpy()
swish_np = swish.numpy()

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(x_np, gelu_np, label='GELU', color='blue')
plt.plot(x_np, relu_np, label='ReLU', color='green')
plt.plot(x_np, swish_np, label='Swish', color='red')

plt.title('Comparacin de funciones de activacin')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

# Crear rango de entrada
x = torch.linspace(-5, 5, steps=1000)

# Activaciones
relu = F.relu(x)
gelu = F.gelu(x)
swish = x * torch.sigmoid(x)
tanh = torch.tanh(x)
leaky_relu = F.leaky_relu(x, negative_slope=0.1)
elu = F.elu(x, alpha=1.0)
sigmoid = torch.sigmoid(x)

# Convertir a numpy
x_np = x.numpy()
relu_np = relu.numpy()
gelu_np = gelu.numpy()
swish_np = swish.numpy()
tanh_np = tanh.numpy()
leaky_relu_np = leaky_relu.numpy()
elu_np = elu.numpy()
sigmoid_np = sigmoid.numpy()

# Graficar
plt.figure(figsize=(12, 7))

plt.plot(x_np, relu_np, label='ReLU', linestyle='-', color='green')
plt.plot(x_np, gelu_np, label='GELU', linestyle='-', color='blue')
plt.plot(x_np, swish_np, label='Swish', linestyle='-.', color='red')
plt.plot(x_np, tanh_np, label='tanh', linestyle=':', color='purple')
plt.plot(x_np, leaky_relu_np, label='Leaky ReLU', linestyle='--', color='orange')
plt.plot(x_np, elu_np, label='ELU', linestyle='-', color='cyan')
plt.plot(x_np, sigmoid_np, label='Sigmoid', linestyle='--', color='brown')

plt.title('Comparacin de funciones de activacin')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
