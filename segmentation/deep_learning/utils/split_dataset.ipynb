```python
import os
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import random
import shutil
```

```python
lidar_path = Path("/home/fmartinez/datasets/goose/lidar/train/")
labels_path = Path("/home/fmartinez/datasets/goose/labels/train/")
```

```python
test_path = Path("/home/fmartinez/datasets_val/labels/val/")

len_test = len(os.listdir(test_path))
print(f"Number of files in test path: {len_test}")
```

```python
list_lidar = sorted(os.listdir(lidar_path))
list_labels = sorted(os.listdir(labels_path))
```

```python

len_lidar = len(list_lidar)
len_labels = len(list_labels)
if len_lidar != len_labels:
    print("Lidar and labels lists are not the same length")
    sys.exit(1)
else:
    print(f"Lidar and labels lists are the same length: {len_lidar} files")
```

```python
total_length = len_test + len_lidar
```

```python
print(total_length)
```

```python
percentage_val = len_test * 100 / total_length
```

```python
print(percentage_val)
```

```python
idxs = random.sample(range(len_labels), len_test)

val_labels = sorted([list_labels[i] for i in idxs])
val_lidar = sorted([list_lidar[i] for i in idxs])
```

```python
val_lidar
```

```python
output_lidar_path = Path("/home/fmartinez/datasets/goose/lidar/val/")
output_labels_path = Path("/home/fmartinez/datasets/goose/labels/val/")

# Crear carpetas de salida si no existen
output_lidar_path.mkdir(parents=True, exist_ok=True)
output_labels_path.mkdir(parents=True, exist_ok=True)

# Mover archivos LiDAR
for nombre_archivo in val_lidar:
    src = lidar_path / nombre_archivo
    dst = output_lidar_path / nombre_archivo
    shutil.move(str(src), str(dst))

# Mover archivos de etiquetas
for nombre_archivo in val_labels:
    src = labels_path / nombre_archivo
    dst = output_labels_path / nombre_archivo
    shutil.move(str(src), str(dst))
```
